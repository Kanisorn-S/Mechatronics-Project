```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(1)

ret, frame = cap.read()
ret
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
cap.release()
```

```python
def take_photo():
  cap = cv2.VideoCapture(0)
  ret, frame = cap.read()
  cv2.imwrite('webcamphoto.jpg', frame)
  cap.release()
```

```python
take_photo()
```
